# Покдключение к БД

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('db.sqlite3')
cur = conn.cursor()
```

```python
df = pd.io.sql.read_sql(
    """
        SELECT name FROM sqlite_master WHERE type='table';
    """,
    conn,
)
df
```

```python
df = pd.io.sql.read_sql(
    """
        SELECT * FROM articles_article;
    """,
    conn,
)
df
```

```python
cur.execute(
    """
        UPDATE articles_article
        SET title = 'ОБНОВЛЕННОЕ Введение в Django Framework'
        WHERE id = 1;
    """
)
conn.commit()
```

```python
cur.execute(
    """
        UPDATE articles_article
        SET text = 'ОБНОВЛЕННОЕ Введение в Django Framework'
        WHERE id = 1;
    """,
)
conn.commit()
```

```python
cur.execute(
    """
        UPDATE articles_article
        SET text = 'ОБНОВЛЕННЫЙ Модели в Django - это единственный и окончательный источник информации о ваших данных. Они содержат основные поля и поведение данных, которые вы храните. Как правило, каждая модель отображается на одну таблицу базы данных. Основы: каждая модель - это класс Python, который является подклассом django.db.models.Model. Каждый атрибут модели представляет поле базы данных. Django предоставляет автоматически сгенерированный API доступа к базе данных. Давайте рассмотрим пример простой модели для хранения статей в блоге.'
        WHERE id = 2;
    """,
)
conn.commit()
```
